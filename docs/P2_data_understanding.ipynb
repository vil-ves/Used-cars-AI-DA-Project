### Phase 2 (Data Understanding)

Projektin data on hankittu Kaggle-palvelun sivuilta https://www.kaggle.com/datasets/ananaymital/us-used-cars-dataset/data

Datasetin omistajana toimii nimimerkillä **AnanayMital** Kaggle-palvelun verkkosivuilla. Datasetti on luotu ja julkaistu 14. päivä syyskuuta, vuonna 2020.

## describes the data format or formats

Datan lähteessä ollaan hyvin lyhyesti selitetty datan alkuperä, mutta lainaten lähteestä, datasetti koostuu yli 3 miljoonasta rivistä, eli yli 3 miljoonaa eri käytettyä tosi elämän autoa. Tiedot hankittiin tuolloin suorittamalla itse tehty indeksointirobotti Cargurus-inventaariossa syyskuussa 2020.

Datan lähteen perusteella, datasta löytyy 66 saraketta, joista 32 kpl on merkkijonoja (string), 10 kpl on totuusarvoihin 'True' tai 'False' perustuvia sarakkeita (boolean), 8 kpl on desimaalilukuisia sarakkeita (decimal) ja 16 kpl on tuntemattomia sarakkeita, mutta niitä tutkitaan seuraavaksi omassa koodi ympäristössä.

Ajettaessa koodin, joka tunnistaa jokaisen sarakkeen tietotyypin, saadaan selville mistä eri tietotyypeistä datasetti todellisuudessa koostuu.

- 43 kpl on object-tietotyyppisiä sarakkeita, jotka suurimmaksi osaksi sisältää yhdistettynä merkkijonoja (string) ja lukuarvoja (float64 / int64). Nämä eivät ole meidän kannalta mieluisia sarake arvoja, sillä niitä on lähes mahdoton pystyä ennustamaan koodissa.

- 15 kpl on float64-tietotyyppisiä sarakkeita, jotka poikkeavat int64 tietotyypistä sillä, että ne sisältävät numeerisen arvonsa lisäksi myös desimaalilukuja tarkemman arvon takaamiseksi.

- 6 kpl on int64-tietotyyppisiä sarakkeita, jotka edellä mainitun selityksen mukaan ovat lukuarvoja, jotka eivät sisällä desimaaliarvoja.

- 2 kpl on boolean-tietotyyppisiä sarakkeita, jotka siis sisältävät totuusarvoja joko 'True' tai 'False'. Dataa tarkasteltua, voidaan havaita myös 'Object'-tietotyypin pitävän sisällään ns. Totuusarvoja. Nämä täytyy ottaa huomioon datan valmistelussa ja muuttamalla kyseiset poikkeamat boolean-tietotyypeiksi.

## amount of data (rows/columns)

Datasetti pitää sisällään yli 3 miljoonaa ainutlaatuista/uniikkia arvoa, eli riviä. Kyseisen määrän vuoksi on datankäsittelyssä otettava testattavaksi vain murto-osa dataa, jotta testaus olisi sujuvampaa.

Sarakkeita datasetissä on myös kohtalaisen paljon, 66 kappaletta. Nämä sarakkeet koostuvat eri tietotyypeistä: object, float64, int64 ja boolean.

## results of data research (including data visualization)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as ticker
```

```python
df = pd.read_csv('../../used_cars_data.csv', nrows=200000, low_memory=False)
df = df.dropna(subset=['mileage', 'horsepower', 'price', 'daysonmarket'])
```

## Data visualization

### Price Distribution

```python
## Poistetaan outlierit
price_90th_percentile = df['price'].quantile(0.90)
print(f"90th Percentile Price: ${price_90th_percentile:.2f}")
```

```python
df[df['price'] <= price_90th_percentile]['price'].hist(bins=50, edgecolor='black')
plt.title('Price Distribution (without Outliers)')
plt.xlabel('Price ($)')
plt.ylabel('Frequency')
plt.show()
```

### Mileage Distribution

```python
df = df.dropna(subset=['mileage'])
mileage_90th_percentile = df['mileage'].quantile(0.90)
print(f"90th Percentile Mileage: {mileage_90th_percentile}")
```

```python
df[df['mileage'] <= mileage_90th_percentile]['mileage'].hist(bins=50, edgecolor='black')
plt.title('Mileage Distribution')
plt.xlabel('Mileage (miles)')
plt.ylabel('Frequency')
plt.grid(False)
plt.show()
```

```python
# autot joiden mailimäärä nolla
zero_mileage_cars = df[df['mileage'] == 0]

# autojen lukumäärä valmistusvuosittain
zero_mileage_years = zero_mileage_cars['year'].value_counts()
print("Model Years of Zero-Mileage Cars:\n", zero_mileage_years)
```

Näyttää siltä että autot joilla ajetut mailimäärä on nolla ovat olleet datan keräysvuonna uusia tai uudehkoja autoja.

 ### Days on Market vs. Price

```python
df['price_bin'] = pd.cut(df['price'], bins=[0, 20000, 40000, 60000, 80000, 100000])

avg_days = df.groupby('price_bin', observed=False)['daysonmarket'].mean()

avg_days.plot(kind='bar', figsize=(8, 6))
plt.title('Average Days on Market by Price Range')
plt.xlabel('Price Range')
plt.ylabel('Average Days on Market')
plt.xticks(rotation=45)
plt.show()
```

Kalliimmat autot näyttäisi olleen pisimpään myynnissä

### Top Manufacturers

```python
# laske Nan arvojen prosenttimäärä
print(df['franchise_make'].isna().sum() / len(df) * 100)
```

```python
# Luo "unknown" kategoria autoilla joiden valmistajaa ei löydy
df['franchise_make'] = df['franchise_make'].fillna("Unknown")

# lasketaan ilmentymät ja piirretään kuvaaja
df['franchise_make'].value_counts().head(10).plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Top 10 Car Manufacturers')
plt.xlabel('Manufacturer')
plt.ylabel('Number of Listings')
plt.xticks(rotation=45)
plt.show()
```

### Price vs horsepower

```python

df.plot.scatter(x='horsepower', y='price', alpha=0.5, color='green')
plt.title('Price vs Horsepower')
plt.xlabel('Horsepower')
plt.ylabel('Price ($)')

plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'${int(x):,}'))

plt.show()
```

### Outlier Analysis

```python
# poistetaan äärimmäiset poikkeavuudet
upper_limit = df['price'].quantile(0.99)
filtered_df = df[df['price'] <= upper_limit]

# luo box plot hinnalle
plt.figure(figsize=(8, 6))
sns.boxplot(x=filtered_df['price'], color='skyblue')

# otsikko ja labelit
plt.title('Outlier Analysis for Price', fontsize=14)
plt.xlabel('Price ($)', fontsize=12)
plt.show()
```

Suurin osa hinnoista sijoittuu 15 000–40 000 dollarin väliin, ja mediaani on noin 30 000 dollaria. Viiksien ulkopuolella olevat harmaat pisteet ovat poikkeamia, tässä ne ovat pääasiassa yli 60 000 dollarin hintoja. Poikkeamat voi olla esimerkiksi luksusautoja.

## Correlation


Sarakkeiden välistä riippuvuutta voi tarkastella erilaisilla korrelaatioilla. Esimerkiksi korrelaatiomatriisi on kätevä tapa etsiä datasta riippuvuuksia eri sarakkeiden väliltä. Korrelaatiomatriisissa näkyy kunkin sarakkeen arvojen korrelaatio kunkin sarakkeen arvon kanssa. Esimerkiksi alla olevassa kuvaajassa maantiekulutus ja kaupunkikulutus näyttävät korreloivan voimakkaasti (0.91), eli korkean kaupunkikulutuksen omaava auto omaa yleensä korkean maantiekulutuksenkin. toisaalta, moottorin iskutilavuus ja taloudellisuus taas korreloivat negatiivisesti, eli iso iskutilavuus aiheuttaa epätaloudellisuutta polttoainekulutukseen.

```python


df_numeric = df.select_dtypes(include=['int', 'float'])
df_numeric = df_numeric.dropna(axis=1, how='all')
df_normalized = (df_numeric - df_numeric.min()) / (df_numeric.max() - df_numeric.min())
correlation_matrix = df_normalized.corr()

plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm',fmt=".2f", vmin=-1, vmax=1)
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.show()
```
